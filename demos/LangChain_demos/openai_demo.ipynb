```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

## DEMO 1 - Basic Usage

```python
# 1) 模型
model = ChatOpenAI(model="gpt-4o-mini", temperature=0.2, api_key = OPENAI_API_KEY)

# 2) 提示模板
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个简洁、专业的助手。"),
    ("human", "请用{style}风格解释：{topic}")
])

# 3) 组装链（LCEL）
chain = prompt | model | StrOutputParser()

# 4) 调用
result = chain.invoke({
    "style": "通俗易懂",
    "topic": "什么是向量数据库"
})

print(result)
```

## DEMO 2 - Tools

```python
from langchain_core.tools import tool
```

```python
@tool
def add(a: int, b: int) -> int:
    """两数相加"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """两数相乘"""
    return a * b

tools = {t.name: t for t in [add, multiply]}

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0, api_key = OPENAI_API_KEY).bind_tools([add, multiply])
ai_msg = llm.invoke("先算 2+3，再把结果乘以 4")

for call in ai_msg.tool_calls:
    name = call["name"]
    args = call["args"]
    result = tools[name].invoke(args)
    print(f"{name}({args}) = {result}")
```
